```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.rcParams["font.sans-serif"] = ["SimHei"]#设置全局变量字体为黑体
plt.rcParams["axes.unicode_minus"] = False#设置全局变量正常显示负号
```

# 柱状图

```python
#some simple data
x = [1,2,3,4,5,6,7,8]
y = np.random.random(8)
#create bar
plt.bar(x,y,width=0.8,color = 'green')
#set x,y_axis label
plt.xlabel('横轴编号')
plt.ylabel('编号对应数量')
plt.savefig(r'C:\Users\Admin\Desktop\新建文件夹\zzt.png',dpi=300)
plt.show()
```

```python
#some simple data
x = [1,2,3,4,5,6,7,8]
y1 = np.random.random(8)
y2 = np.random.random(8)
#create bar
plt.bar(x,y1,width=0.8,color = 'green')
plt.bar(x,y2,width=0.8,color = 'yellow',bottom = y1)
#set x,y_axis label
plt.xlabel('横轴编号')
plt.ylabel('编号对应数量')
plt.savefig(r'C:\Users\Admin\Desktop\新建文件夹\ddzzt.png',dpi=300)
plt.show()
```

```python
#some simple data
boxweight = np.random.randint(0,10,100)
x = boxweight
bins = range(0,11,1)
plt.hist(x,bins = bins ,color = 'green',histtype='bar',edgecolor = 'b')
#set x,y_axis label
plt.xlabel('横轴编号')
plt.ylabel('编号对应数量')
plt.savefig(r'C:\Users\Admin\Desktop\新建文件夹\zft.png',dpi=300)
plt.show()
```

```python
#some simple data
x = np.random.randn(1000)
#create boxplot
plt.boxplot(x,patch_artist = True,meanline = False,showmeans = True,showfliers = True)
#set x_axis stick and y_axis label
plt.xticks([1],['随机数生成器AlphaRM'])
plt.ylabel('随即数量')
plt.savefig(r'C:\Users\Admin\Desktop\新建文件夹\xxt.png',dpi=300)
plt.show()
```

```python
#some simple data
x = np.random.rand(50)
y = np.random.random(50)
#create scatter
plt.scatter(x,y,color = 'green')
#set x,y_axis label
plt.xlabel('横轴编号')
plt.ylabel('编号对应数量')
plt.savefig(r'C:\Users\Admin\Desktop\新建文件夹\sdt.png',dpi=300)
plt.show()
```

```python
#some simple data
x = [1,2,3,4,5,6,7,8]
y = np.random.random(8)
#create plot
plt.plot(x,y,color = 'green',marker="o")
#set x,y_axis label
plt.xlabel('横轴编号')
plt.ylabel('编号对应数量')
plt.savefig(r'C:\Users\Admin\Desktop\新建文件夹\zxt.png',dpi=300)
plt.show()
```

```python
#some simple data
x = [1,2,3,4,5,6,7,8]
y = np.random.random(8)
#create plot
plt.plot(x,y,color = 'green',marker="o",label="图例")
#set x,y_axis label
plt.xlabel('横轴编号')
plt.ylabel('y轴标签')
plt.xlabel('x轴标签')
plt.title("图表标题")
plt.grid()
plt.legend()
plt.show()
```

```python
columns = ['A', 'B', 'C', 'D']
rows = ['1', '2', '3', '4'] 
data = np.random.random((4,4))
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(data, cmap=plt.cm.hot)
fig.colorbar(cax)
ax.set_xticklabels([''] + columns)
ax.set_yticklabels([''] + rows)
plt.savefig(r'C:\Users\Admin\Desktop\新建文件夹\rlt.png',dpi=300)
plt.show()
```
