# 导包

```python
import pandas as pd
from dbfread import DBF
```

```python
csv_df = pd.read_csv(r'./使用数据/样木库.csv',encoding='utf_8_sig')
csv_df
```

```python
excel_df = pd.read_excel(r'./使用数据/属性表.xls')
sam_excel_df = excel_df.sample(10).sample(10,axis=1)
sam_excel_df
```

```python
table = DBF('./使用数据/湖北样木库T2004423.DBF',encoding='gbk',char_decode_errors='ignore',load=True)
```

```python
dbf_df = pd.DataFrame(iter(table))
sam_dbf_df = dbf_df.sample(10).sample(10,axis=1)
sam_dbf_df
```

```python
csv_df.to_csv(r'./输出数据/样木库.csv',encoding='utf_8_sig',index=False)
```

```python
sam_excel_df.to_excel(r'./输出数据/属性表.xls',index=False)
```

```python
sam_dbf_df.to_excel(r'./输出数据/dbf_样木库.xls',index=False)
```
