```python
import pandas as pd
import numpy as np

file_xls = r"./使用数据/属性表.xls"
df = pd.DataFrame(pd.read_excel(file_xls))

newdf = pd.DataFrame(df, columns=["土地所有权","森林类别","国家级公益林原保护等级","地类","面积亩" ])
newdf
```

```python
grouped = newdf.groupby(["土地所有权","森林类别","国家级公益林原保护等级","地类"],as_index = False)
grouped.size()
```

```python
grouped1 = newdf.groupby(["土地所有权","森林类别","国家级公益林原保护等级","地类"])["面积亩"]
grouped1.size()
```
