```python
import pandas as pd
import numpy as np
import os
```

```python
data = pd.read_excel(r'./使用数据/属性表.xls')
df1 = data[['乡镇场','林木使用权','森林类别','地类','面积亩']]#林地资源现状
df1['面积公顷']=df1['面积亩']/15
```

```python
print(list(data))
```

各类土地面积统计表

```python
grouped_df = df1.groupby(['乡镇场','林木使用权','森林类别','地类']).agg({'面积公顷':np.sum})
grouped_df=grouped_df.unstack()
grouped_df
```

```python
grouped_df.to_excel(r'./输出数据/各类土地面积统计表.xls')
```
